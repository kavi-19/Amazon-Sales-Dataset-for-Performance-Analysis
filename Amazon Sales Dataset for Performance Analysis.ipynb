```python
df['Product Name']=df['Product Name'].fillna('Unknow')
```

```python
# Convert 'Order Date' to datetime with day first
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

# Check the result
df['Order Date'].head()
```

```python
df['Order Month'] = df['Order Date'].dt.month
df['Order Year'] = df['Order Date'].dt.year
```

```python
df.head()
```

```python
# Aggregations for Analysis
# Total sales by Region
sales_by_region = df.groupby('Region')['Total Sales'].sum().sort_values(ascending=False)
print(sales_by_region)

# Total sales by Product Category
sales_by_category = df.groupby('Product Category')['Total Sales'].sum().sort_values(ascending=False)
print(sales_by_category)

# Top 10 products by Total Sales
top_products = df.groupby('Product Name')['Total Sales'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sales by Region
plt.figure(figsize=(8,5))
sns.barplot(x=sales_by_region.index, y=sales_by_region.values)
plt.title("Total Sales by Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Sales by Product Category
plt.figure(figsize=(8,5))
sns.barplot(x=sales_by_category.index, y=sales_by_category.values)
plt.title("Total Sales by Product Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 Products by Sales
plt.figure(figsize=(10,6))
sns.barplot(x=top_products.index, y=top_products.values)
plt.title("Top 10 Products by Sales")
plt.xticks(rotation=90)
plt.show()
```

```python
## KPI
total_sales = df['Total Sales'].sum()
total_quantity = df['Quantity Sold'].sum()
average_discount = df['Discount (%)'].mean()
average_profit_margin = df['Profit Margin'].mean()

print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Units Sold: {total_quantity}")
print(f"Average Discount: {average_discount:.2f}%")
print(f"Average Profit Margin: {average_profit_margin:.2f}%")
```

```python
##Total Sales over time
sales_over_time = df.groupby('Order Date')['Total Sales'].sum()
plt.figure(figsize=(12,5))
sales_over_time.plot()
plt.title("Total Sales Over Time")
plt.xlabel("Order Date")
plt.ylabel("Total Sales")
plt.show()
```

```python
##outliers
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Total Sales'])
plt.title("Boxplot for Total Sales")
plt.show()
```

```python
# Example for 'Total Sales'
Q1 = df['Total Sales'].quantile(0.25)
Q3 = df['Total Sales'].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = df[(df['Total Sales'] < (Q1 - 1.5 * IQR)) | (df['Total Sales'] > (Q3 + 1.5 * IQR))]
print("Number of outliers in Total Sales:", len(outliers))
```

```python
##Cap/Clamp Outliers
upper_limit = Q3 + 1.5 * IQR
lower_limit = Q1 - 1.5 * IQR
df['Total Sales'] = df['Total Sales'].clip(lower=lower_limit, upper=upper_limit)
```
